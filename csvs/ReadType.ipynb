```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib 

font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 40}

matplotlib.rc('font', **font)
```

```python
def create_dic(dataset):
    data = {}
    for i in range(0, len(dataset)):
        data[dataset.loc[i][0]] = list(map(float, dataset.loc[i][1][1:-1].split(',')))
    return data

   
# Read in the files
data0 = pd.read_csv("type_PSO_prob_0_output.csv", header=None)
data0 = create_dic(data0)

data1 = pd.read_csv("type_GSA_prob_0_output.csv", header=None)
data1 = create_dic(data1)

data2 = pd.read_csv("type_RAN_prob_0_output.csv", header=None)
data2 = create_dic(data2)

keys = list(data1.keys())
```

```python
plt.rc('ytick',labelsize=100)
fig, axs = plt.subplots(5, 4)
col = 0
row = 0
fig.set_size_inches(100, 100)
plt.yticks(fontsize=100)
plt.title("Convex Optimization")


for i in range(len(data1)):
    axs[row, col].set_title(keys[i] , fontsize=100)
    overall = [data0[keys[i]], data1[keys[i]]]
    axs[row][col].violinplot(overall)
    xticklabels = ["PSO", "GSA"]
    axs[row][col].set_xticks([1,2,3])
    axs[row][col].set_xticklabels(xticklabels,fontsize=100)
    row = row + 1
    if row == 5:
        row =0
        col = col + 1
plt.tight_layout()
```
