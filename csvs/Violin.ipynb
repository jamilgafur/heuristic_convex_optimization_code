```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib 

font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 40}

matplotlib.rc('font', **font)

def getK_N(data, row):
    val = data.iloc[row][:2].tolist()
    return "n: " + str(val[0]) + " k:" + str(val[1])

def getn_m_M_b(data, row):
    val = data.iloc[row][:5].tolist()
    return "n: " + str(val[0]) + " m:" + str(val[1])+ " M:" + str(val[2])+ " b:" + str(val[3])
```

```python
data0 = pd.read_csv("GA_seed_1234_prob_1_all.csv")
data1 = pd.read_csv("GSA_seed_1234_prob_1_all.csv")
data2 = pd.read_csv("PSO_seed_1234_prob_1_all.csv")
data3 = pd.read_csv("RAN_seed_1234_prob_1_all.csv")
plt.rc('ytick',labelsize=100)
fig, axs = plt.subplots(5, 4)
col = 0
row = 0
fig.set_size_inches(100, 100)
plt.yticks(fontsize=100)
plt.title("Convex Optimization")
for i in range(len(data0)):
    axs[row, col].set_title(getK_N(data0, i), fontsize=100)
    overall = list([data0.iloc[i][2:].tolist(), data1.iloc[i][2:].tolist(), data2.iloc[i][2:].tolist(), data3.iloc[i][2:].tolist()])
    parts = axs[row][col].violinplot(overall, showmeans=True, showmedians=True, showextrema=True)
    xticklabels = ["GA", "GSA", "PSO", "RAN"]
    axs[row][col].set_xticks([1,2,3,4])
    axs[row][col].set_xticklabels(xticklabels,fontsize=100)
    axs[row][col].set_yscale('log')#set_ylim([0,4])
    row = row + 1
    if row == 5:
        row =0
        col = col + 1
plt.tight_layout()

plt.savefig("Violin plot problem 1", dpi=100)
plt.show()
```

```python
data0 = pd.read_csv("GA_seed_1234_prob_2_all.csv")
data1 = pd.read_csv("GSA_seed_1234_prob_2_all.csv")
data2 = pd.read_csv("PSO_seed_1234_prob_2_all.csv")
data3 = pd.read_csv("RAN_seed_1234_prob_2_all.csv")
plt.rc('ytick',labelsize=100)
fig, axs = plt.subplots(5, 4)
col = 0
row = 0
fig.set_size_inches(100, 100)
plt.yticks(fontsize=100)

for i in range(10, len(data0), 5):
    axs[row, col].set_title(getn_m_M_b(data0, i), fontsize=100)
    overall = list([data0.iloc[i][4:].tolist(), data1.iloc[i][4:].tolist(), data2.iloc[i][4:].tolist(), data3.iloc[i][4:].tolist()])
    axs[row][col].violinplot(overall,showmeans=False, showmedians=True, showextrema=False)
    xticklabels = ["GA", "GSA", "PSO", "RAN"]
    axs[row][col].set_xticks([1,2,3,4])
    axs[row][col].set_xticklabels(xticklabels,fontsize=100)
    row = row + 1
    if row == 5:
        row =0
        col = col + 1
plt.tight_layout()

plt.savefig("Violin plot problem 2", dpi=100)
plt.show()
```
