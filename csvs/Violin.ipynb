```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib 

font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 40}

matplotlib.rc('font', **font)

def getK_N(data, row):
    val = data.iloc[row][:2].tolist()
    return "n: " + str(val[0]) + " k:" + str(val[1])
```

```python
data0 = pd.read_csv("GSA_seed_1234_prob_1_all.csv")
data1 = pd.read_csv("GSA_seed_1234_prob_1_all.csv")
data2 = pd.read_csv("PSO_seed_1234_prob_1_all.csv")
data3 = pd.read_csv("RAN_seed_1234_prob_1_all.csv")

fig, axs = plt.subplots(5,4)
col = 0
row = 0

fig.set_size_inches(100, 100)

for i in range(len(data0)):
    axs[row, col].set_title(getK_N(data0, i), fontsize=40)
    overall = list([data0.iloc[i][2:].tolist(), data1.iloc[i][2:].tolist(), data2.iloc[i][2:].tolist(), data3.iloc[i][2:].tolist()])
    axs[row][col].violinplot(overall)
    xticklabels = ["GA", "GSA", "PSO", "RAN"]
    axs[row][col].set_xticks([1,2,3,4])
    axs[row][col].set_xticklabels(xticklabels,fontsize=40)
    #axs[row][col].set_yscale('log')
    row = row + 1
    if row == 5:
        row =0
        col = col + 1
plt.tight_layout()

plt.savefig("Violin plot problem 1")
plt.show()
```
