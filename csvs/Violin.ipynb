```python
import matplotlib.pyplot as plt
import pandas as pd

def getK_N(data, row):
    val = data.iloc[row][:2].tolist()
    return "n: " + str(val[0]) + " k:" + str(val[1])
```

```python
data = pd.read_csv("PSO_seed_1234_prob_1_all.csv")
```

```python
data.head()
```

```python
for i in range(len(data)):
    plt.violinplot(data.iloc[i][2:].tolist(), showextrema=True, showmeans=True)
    plt.title(getK_N(data,i))
    plt.show()
```
